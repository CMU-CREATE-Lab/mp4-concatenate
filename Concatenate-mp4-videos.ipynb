```python
import pprint, struct, urllib
```

```python
class MP4:
    def __init__(self, filename):
        self.fp = open(filename, 'r')
        self.info = self.parse_container()
        
    def read32(self):
        return struct.unpack('!I', self.fp.read(4))[0]

    def read16(self):
        return struct.unpack('!H', self.fp.read(2))[0]

    def parse_elst(self):
        version = self.fp.read(1)
        flags = self.fp.read(3)
        count = self.read32()
        ret = []
        for i in range(0, count):
            elt = {}
            elt['duration'] = self.read32()
            elt['start_time'] = self.read32()
            elt['rate'] = self.read32() / 65536.0
            print '  elst element %s' % elt
            ret.append(elt)
        return ret

    def parse_mvhd(self):
        version = self.fp.read(1)
        flags = self.fp.read(3)
        ret = {}
        ret['creation_time'] = self.read32()
        ret['modification_time'] = self.read32()
        ret['time_scale'] = self.read32()
        ret['duration'] = float(self.read32()) / ret['time_scale']
        print '  mvhd Movie Header: %s' % ret
        return ret

    def parse_tkhd(self):
        version = self.fp.read(1)
        flags = self.fp.read(3)
        ret = {}
        ret['creation_time'] = self.read32()
        ret['modification_time'] = self.read32()
        ret['track_id'] = self.read32()
        self.fp.read(4)
        ret['duration'] = float(self.read32()) # needs to be scaled by mvhd['time_scale']
        self.fp.read(8 + 2 + 2 + 2 + 2 + 36)
        ret['track_width'] = self.read32() / 65536.0
        ret['track_height'] = self.read32() / 65536.0
        print '  tkhd Track Header: %s' % ret
        return ret

    def parse_mdhd(self):
        version = self.fp.read(1)
        flags = self.fp.read(3)
        ret = {}
        ret['creation_time'] = self.read32()
        ret['modification_time'] = self.read32()
        ret['time_scale'] = self.read32()
        ret['duration'] = float(self.read32()) / ret['time_scale']
        return ret
    
    def parse_stsd(self):
        version = self.fp.read(1)
        flags = self.fp.read(3)
        num = self.read32()
        ret = []
        for i in range(0, num):
            desc = {}
            pos = self.fp.tell()
            size = self.read32()
            desc['format'] = self.fp.read(4)
            self.fp.read(6)
            desc['index'] = self.read16()
            self.fp.seek(pos + size)
            ret.append(desc)
        return ret

    # Chunk offset table
    def parse_stco(self):
        version = self.fp.read(1)
        flags = self.fp.read(3)
        num = self.read32()
        ret = []
        for i in range(0, num):
            ret.append({'chunk_offset': self.read32()})
        return ret

    def parse_stsz(self):
        version = self.fp.read(1)
        flags = self.fp.read(3)
        sample_size = self.read32()
        if sample_size != 0:
            raise Exception('sample_size of != 0 is unimplemented' % sample_size)
        num = self.read32()
        sample_sizes = []
        for i in range(0, num):
            sample_sizes.append(self.read32())
        return {'samples_sizes': sample_sizes}
        
    def parse_stsc(self):
        version = self.fp.read(1)
        flags = self.fp.read(3)
        num = self.read32()
        ret = []
        for i in range(0, num):
            entry = {}
            entry['first_chunk'] = self.read32()
            entry['samples_per_chunk'] = self.read32()
            entry['sample_description_id'] = self.read32()
            ret.append(entry)
        return ret

    def parse_stss(self):
        version = self.fp.read(1)
        flags = self.fp.read(3)
        num = self.read32()
        key_frame_samples = []
        for i in range(0, num):
            key_frame_samples.append(self.read32())
        return {'key_frame_samples': key_frame_samples}

    def parse_stts(self):
        version = self.fp.read(1)
        flags = self.fp.read(3)
        num = self.read32()
        ret = []
        for i in range(0, num):
            entry = {}
            entry['sample_count'] = self.read32()
            entry['sample_duration'] = self.read32() # scale using mdhd.time_scale
            ret.append(entry)
        return ret

    def parse_container(self, offset0=0, offset1=None, prefix=''):
        "Walk the atom tree in a mp4 file"
        if offset1 == None:
            self.fp.seek(0, 2)
            offset1 = self.fp.tell()
        offset= offset0
        ret = {}
        while offset < offset1:
            self.fp.seek(offset)
            atomsize= struct.unpack("!i", self.fp.read(4))[0]
            atomtype= self.fp.read(4)
            parser_name = 'parse_' + atomtype
            if parser_name in dir(self):
                print 'Found %s size %d' % (prefix + atomtype, atomsize)
                val = getattr(self, parser_name)()
            elif atomtype in ['moov', 'trak', 'mdia', 'minf', 'edts', 'dinf',
                            'stbl', 'udta']:
                print 'Found %s size %d' % (prefix + atomtype, atomsize)
                val = self.parse_container(offset + 8, offset + atomsize, prefix + atomtype + '.')
            elif atomtype in ['meta']:
                print 'Found %s size %d' % (prefix + atomtype, atomsize)
                val = self.parse_container(offset + 12, offset + atomsize, prefix + atomtype + '.')
            elif atomtype in ['ftyp', 'mdhd', 'hdlr',
                              'vmhd', 'dref', 'stsd', 'stts', 'stss', 'stsc',
                              'stsz', 'stco', 'ilst', 'free']:
                val = None
                print 'Found %s size %d' % (prefix + atomtype, atomsize)
            else:
                snippet = self.fp.read(min(atomsize - 8, 100))
                print 'Found %s size %d: "%s"' % (prefix + atomtype, atomsize, urllib.quote(snippet))
                val = None
            ret[atomtype] = val
            offset += atomsize
        return ret

m = MP4('short.mp4')
pprint.pprint(m.info)
```
