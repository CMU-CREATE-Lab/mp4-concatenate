-------------------------------

```python
import copy, json, pprint, struct, urllib
```

```python
class MP4:
    def __init__(self, filename):
        self.fp = open(filename, 'r')
        self.verbose = False
        self.info = self.parse_container()
        
    def read32(self):
        return struct.unpack('!I', self.fp.read(4))[0]

    def read16(self):
        return struct.unpack('!H', self.fp.read(2))[0]
    
    def write32(self, out, val):
        out.write(struct.pack('!I', val))

    def parse_elst(self):
        version = self.fp.read(1)
        flags = self.fp.read(3)
        count = self.read32()
        ret = []
        for i in range(0, count):
            elt = {}
            elt['duration'] = self.read32()
            elt['start_time'] = self.read32()
            elt['rate'] = self.read32() / 65536.0
            if self.verbose:
                print '  elst element %s' % elt
            ret.append(elt)
        return {'edits': ret}

    def parse_mvhd(self):
        version = self.fp.read(1)
        flags = self.fp.read(3)
        ret = {}
        ret['creation_time'] = self.read32()
        ret['modification_time'] = self.read32()
        ret['time_scale'] = self.read32()
        ret['duration'] = float(self.read32()) / ret['time_scale']
        if self.verbose:
            print '  mvhd Movie Header: %s' % ret
        return ret

    def parse_tkhd(self):
        version = self.fp.read(1)
        flags = self.fp.read(3)
        ret = {}
        ret['creation_time'] = self.read32()
        ret['modification_time'] = self.read32()
        ret['track_id'] = self.read32()
        self.fp.read(4)
        ret['duration'] = float(self.read32()) # needs to be scaled by mvhd['time_scale']
        self.fp.read(8 + 2 + 2 + 2 + 2 + 36)
        ret['track_width'] = self.read32() / 65536.0
        ret['track_height'] = self.read32() / 65536.0
        if self.verbose:
            print '  tkhd Track Header: %s' % ret
        return ret

    def parse_mdhd(self):
        version = self.fp.read(1)
        flags = self.fp.read(3)
        ret = {}
        ret['creation_time'] = self.read32()
        ret['modification_time'] = self.read32()
        ret['time_scale'] = self.read32()
        ret['duration'] = float(self.read32()) / ret['time_scale']
        return ret
    
    def parse_stsd(self):
        version = self.fp.read(1)
        flags = self.fp.read(3)
        num = self.read32()
        ret = []
        for i in range(0, num):
            desc = {}
            pos = self.fp.tell()
            size = self.read32()
            desc['format'] = self.fp.read(4)
            self.fp.read(6)
            desc['index'] = self.read16()
            self.fp.seek(pos + size)
            ret.append(desc)
        return {'sample_descriptions': ret}

    # Chunk offset table
    def parse_stco(self):
        version = self.fp.read(1)
        flags = self.fp.read(3)
        num = self.read32()
        ret = []
        for i in range(0, num):
            ret.append(self.read32())
        return {'chunk_offsets': ret}

    # Chunk offset table
    def write_stco(self, out, atom):
        chunk_offsets = atom['chunk_offsets']
        self.write32(out, 16 + 4 * len(chunk_offsets))
        out.write('stco')
        out.write('\x00' * 1) # version
        out.write('\x00' * 3) # flags
        self.write32(out, len(chunk_offsets))
        for offset in chunk_offsets:
            self.write32(out, offset)
            
    def parse_stsz(self):
        version = self.fp.read(1)
        flags = self.fp.read(3)
        sample_size = self.read32()
        if sample_size != 0:
            raise Exception('sample_size of != 0 is unimplemented' % sample_size)
        num = self.read32()
        sample_sizes = []
        for i in range(0, num):
            sample_sizes.append(self.read32())
        return {'samples_sizes': sample_sizes}
        
    def parse_stsc(self):
        version = self.fp.read(1)
        flags = self.fp.read(3)
        num = self.read32()
        ret = []
        for i in range(0, num):
            entry = {}
            entry['first_chunk'] = self.read32()
            entry['samples_per_chunk'] = self.read32()
            entry['sample_description_id'] = self.read32()
            ret.append(entry)
        return {'sample_to_chunk_map': ret}

    def parse_stss(self):
        version = self.fp.read(1)
        flags = self.fp.read(3)
        num = self.read32()
        key_frame_samples = []
        for i in range(0, num):
            key_frame_samples.append(self.read32())
        return {'key_frame_samples': key_frame_samples}

    def parse_stts(self):
        version = self.fp.read(1)
        flags = self.fp.read(3)
        num = self.read32()
        ret = []
        for i in range(0, num):
            entry = {}
            entry['sample_count'] = self.read32()
            entry['sample_duration'] = self.read32() # scale using mdhd.time_scale
            ret.append(entry)
        return {'time_to_sample_map': ret}
    
    def parse_container(self, offset0=0, offset1=None, prefix=''):
        "Walk the atom tree in a mp4 file"
        if offset1 == None:
            self.fp.seek(0, 2)
            offset1 = self.fp.tell()
        offset= offset0
        ret = OrderedDict()
        while offset < offset1:
            self.fp.seek(offset)
            atomsize= struct.unpack("!i", self.fp.read(4))[0]
            atomtype= self.fp.read(4)
            parser_name = 'parse_' + atomtype
            if parser_name in dir(self):
                if self.verbose:
                    print 'Found %s size %d' % (prefix + atomtype, atomsize)
                val = getattr(self, parser_name)()
            elif atomtype in ['meta', 'moov', 'trak', 'mdia', 'minf', 'edts', 'dinf',
                            'stbl', 'udta']:
                if self.verbose:
                    print 'Found %s size %d' % (prefix + atomtype, atomsize)
                container_header_size = 8
                if atomtype == 'meta':
                    container_header_size = 12
                val = self.parse_container(offset + container_header_size, offset + atomsize, prefix + atomtype + '.')
                val['_container_header_size'] = container_header_size
            elif atomtype in ['ftyp', 'mdhd', 'hdlr', 'mdat',
                              'vmhd', 'dref', 'stsd', 'stts', 'stss', 'stsc',
                              'stsz', 'stco', 'ilst', 'free']:
                val = {}
                if self.verbose:
                    print 'Found %s size %d' % (prefix + atomtype, atomsize)
            else:
                raise Exception('Unknown atom type "%s"' % atomtype)
            val['_position'] = offset
            val['atomtype'] = atomtype
            val['atomsize'] = atomsize
            ret[atomtype] = val
            offset += atomsize
        return ret

    def copy_with_padding(self, dest, padding):

        moov = copy.deepcopy(self.info['moov'])
        new_mdat_location = (self.info['ftyp']['atomsize'] +
                             self.info['moov']['atomsize'] +
                             8 +
                             padding)
        mdat_move = new_mdat_location - self.info['mdat']['_position']
        print 'Moving mdat by %d bytes' % mdat_move
        stco = moov['trak']['mdia']['minf']['stbl']['stco']
        chunk_offsets = stco['chunk_offsets']
        for i in range(0, len(chunk_offsets)):
            chunk_offsets[i] += mdat_move
                
        with open(dest, 'w') as out:
            self.write_atom(out, self.info['ftyp'])
            self.write_atom(out, moov)
            self.write_free_atom(out, padding)
            self.write_atom(out, self.info['mdat'])
    
    def write_atom(self, out, atom):
        outpos = out.tell()
        writer_name = 'write_' + atom['atomtype']
        if writer_name in dir(self):
            getattr(self, writer_name)(out, atom)
        elif '_container_header_size' in atom:
            self.fp.seek(atom['_position'])
            out.write(self.fp.read(atom['_container_header_size']))
            for (child_type, child_atom) in atom.iteritems():
                if len(child_type) == 4:
                    self.write_atom(out, child_atom)
        else:
            self.fp.seek(atom['_position'])
            out.write(self.fp.read(atom['atomsize']))
        written = out.tell() - outpos
        if written != atom['atomsize']:
            print atom
            raise Exception('Wrote %d bytes for %s but should have written %d' % 
                            (written, atom['atomtype'], atom['atomsize']))
    
    def write_free_atom(self, out, space):
        self.write32(out, space + 8)
        out.write('free')
        out.write('\x00' * space)
```
